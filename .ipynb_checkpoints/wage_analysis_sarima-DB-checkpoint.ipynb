## Running SARIMA models for wages data

This notebook will plot the moving average component of the wages data for each of the localities that were analyzed. These plots were analyzed qualitatively to assess if there was any discernable impact resulting from Hurricane Sandy.

```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt

import statsmodels.api as sm
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose


from statsmodels.tsa.stattools import adfuller

from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
# Code written by Joseph Nelson. This is used to perform the augmented Dickey-Fuller test.

def interpret_dftest(dftest):
    dfoutput = pd.Series(dftest[0:2], index=['Test Statistic','p-value'])
    return dfoutput
```

```python
atlantic_wage = pd.read_csv('./bls-api-data/AtlanticCity2010-2015.csv')
boston_wage = pd.read_csv('./bls-api-data/Boston2010-2015.csv')
bridgeport_wage = pd.read_csv('./bls-api-data/Bridgeport2010-2015.csv')
new_haven_wage = pd.read_csv('./bls-api-data/NewHaven2010-2015.csv')
new_london_wage = pd.read_csv('./bls-api-data/NewLondon2010-2015.csv')
newport_wage = pd.read_csv('./bls-api-data/Newport2010-2015.csv')
baltimore_wage = pd.read_csv('./bls-api-data/Baltimore2010-2015.csv')
ocean_city_wage = pd.read_csv('./bls-api-data/WorcesterCounty2010-2015.csv')
portland_wage = pd.read_csv('./bls-api-data/Portland2010-2015.csv')
providence_wage = pd.read_csv('./bls-api-data/Providence2010-2015.csv')
```

```python
def clean_wage(df):
    df.rename(columns={'date': 'Date'}, inplace=True)
    df.drop('Unnamed: 0', axis=1, inplace=True)
    #df['Date'] = pd.to_datetime(df['Date'])
    df = df.set_index('Date')
    df.index = pd.to_datetime(df.index)
    return df
```

```python
atlantic_wage = clean_wage(atlantic_wage)
boston_wage = clean_wage(boston_wage)
bridgeport_wage = clean_wage(bridgeport_wage)
new_haven_wage = clean_wage(new_haven_wage)
new_london_wage = clean_wage(new_london_wage)
newport_wage = clean_wage(newport_wage)
baltimore_wage = clean_wage(baltimore_wage)
ocean_city_wage = clean_wage(ocean_city_wage)
portland_wage = clean_wage(portland_wage)
providence_wage = clean_wage(providence_wage)
```

The function below will be used to generate SARIMA models for the wage data. These models take one order of seasonal differencing. We are honing in on the moving average component of these time series in order to discern any sudden fluctuations that may have occurred around the time of Hurricane Sandy.

```python
def run_sarima(df, col):

    # Iterate through all possible differences.
    for d in range(1, len(df)):

        # Print out a counter and the corresponding p-value.
        print(f'Checking difference of {d}.')
        print(f'p-value = {interpret_dftest(adfuller(df[col].diff(12).diff(d).dropna()))["p-value"]}.')

        # If our data, differenced by d time periods, are stationary, print that out!
        if interpret_dftest(adfuller(df[col].diff(12).diff(d).dropna()))['p-value'] < 0.05:
            print(f'Differencing our time series by d={d} yields a stationary time series!')
            break

        print()

    model = SARIMAX(endog = df[col].dropna(), # endog = Y variable
                  order = (0,1,1), seasonal_order=(0, 1, 1, 12), enforce_invertibility=False) # (p, d, q) <-- talk about q later!

    model_fit = model.fit()
    y_hat = model_fit.predict()
              
    plt.figure() 
    plt.title(col)
    plt.plot(y_hat);
          
          
```

```python
for col in wage_cols:
    run_sarima(atlantic_wage, col)
```

## Takeaways

In general, this wage data seems to be faitly static. It seems to be only influenced by seasonal trends and long term trends. It is hard to imagine how any impact of an isolated occurrence like a natural disaster would be discerned. Perhaps it could be done if we had significantly longer time spans of data and data that was complete. The main takeaway would be that it seems like wages is not a good metric to analyze in order to discern sudden fluctuations in the economy, since wages are generally steady. Below, we will use Atlantic City as an example and show specific plots from that time series. The wage patterns were similar across localities.

```python
def plot_leisure_model(df, city):

    model = SARIMAX(endog = df["leisure_hospitality_aggr_value"].dropna(), # endog = Y variable
                  order = (0,1,1), seasonal_order=(0, 1, 1, 12), enforce_invertibility=False) # (p, d, q) <-- talk about q later!

    model_fit = model.fit()
    y_hat = model_fit.predict()
              
    plt.figure(figsize = (25, 15)) 
    plt.title(f"{city}", fontsize=60, pad=20)
    plt.xticks(fontsize = 40)
    plt.yticks(fontsize = 40)
    plt.ylabel("Hundreds of Millions of Dollars", fontsize = 40, labelpad=20)
    plt.tight_layout()
    plt.plot(y_hat, linewidth = 6);
    
```

```python
plot_leisure_model(atlantic_wage, "Atlantic City")
```
