```python
def clean_baro(df):
    #df['DATE TIME'] = df['DATE TIME']#.map(lambda x: x[:10])
    df = df[['DATE TIME', ' BARO']]
    df.rename(columns={'DATE TIME': 'Date', ' BARO': 'baro'}, inplace=True)
    #df['Date'] = df['Date'].map(lambda x: x[:7])
    df.dropna(inplace=True)
    df = df.set_index('Date')
    df.index = pd.to_datetime(df.index)#, format='%Y-%m')
    df['baro'] = pd.to_numeric(df['baro'])
    return df
```

```python
atlantic_baro = pd.concat([baro_2010_atlantic, baro_2011_atlantic, baro_2012_atlantic, baro_2013_atlantic, baro_2014_atlantic, baro_2015_atlantic])
boston_baro = pd.concat([baro_2010_boston, baro_2011_boston, baro_2012_boston, baro_2013_boston, baro_2014_boston, baro_2015_boston])
bridgeport_baro = pd.concat([baro_2010_bridgeport, baro_2011_bridgeport, baro_2012_bridgeport, baro_2013_bridgeport, baro_2014_bridgeport, baro_2015_bridgeport])
new_haven_baro = pd.concat([baro_2010_new_haven, baro_2011_new_haven, baro_2012_new_haven, baro_2013_new_haven, baro_2014_new_haven, baro_2015_new_haven])
new_london_baro = pd.concat([baro_2010_new_london, baro_2011_new_london, baro_2012_new_london, baro_2013_new_london, baro_2014_new_london, baro_2015_new_london])
newport_baro = pd.concat([baro_2010_newport, baro_2011_newport, baro_2012_newport, baro_2013_newport, baro_2014_newport, baro_2015_newport])
baltimore_baro = pd.concat([baro_2010_baltimore, baro_2011_baltimore, baro_2012_baltimore, baro_2013_baltimore, baro_2014_baltimore, baro_2015_baltimore])
ocean_city_baro = pd.concat([baro_2010_ocean_city, baro_2011_ocean_city, baro_2012_ocean_city, baro_2013_ocean_city, baro_2014_ocean_city, baro_2015_ocean_city])
portland_baro = pd.concat([baro_2010_portland, baro_2011_portland, baro_2012_portland, baro_2013_portland, baro_2014_portland, baro_2015_portland])
providence_baro = pd.concat([baro_2010_providence, baro_2011_providence, baro_2012_providence, baro_2013_providence, baro_2014_providence, baro_2015_providence])
```

```python
atlantic_baro.to_csv("./weather-clean-data-original-time/atlantic_baro.csv")
boston_baro.to_csv("./weather-clean-data-original-time/boston_baro.csv")
bridgeport_baro.to_csv("./weather-clean-data-original-time/bridgeport_baro.csv")
new_haven_baro.to_csv("./weather-clean-data-original-time/new_haven_baro.csv")
new_london_baro.to_csv("./weather-clean-data-original-time/new_london_baro.csv")
newport_baro.to_csv("./weather-clean-data-original-time/newport_baro.csv")
baltimore_baro.to_csv("./weather-clean-data-original-time/baltimore_baro.csv")
ocean_city_baro.to_csv("./weather-clean-data-original-time/ocean_city_baro.csv")
portland_baro.to_csv("./weather-clean-data-original-time/portland_baro.csv")
providence_baro.to_csv("./weather-clean-data-original-time/providence_baro.csv")
```

```python
cumm = pd.concat([atlantic_baro, boston_baro, bridgeport_baro, 
                  baltimore_baro, new_haven_baro, new_london_baro,
                  newport_baro, baltimore_baro,
                  ocean_city_baro,portland_baro, providence_baro
                  ], 
                 axis=0)
```

## The figure below visualizes the global air pressure average (March,2011-December,2015)


```python
plt.figure(figsize=(25, 15))
plt.plot(atlantic_baro, label='Atlantic City')
plt.plot(boston_baro, label='Boston')
plt.plot(bridgeport_baro, label='Bridgeport')
plt.plot(new_haven_baro, label='New Haven')
plt.plot(new_london_baro, label='New London')
plt.plot(newport_baro, label='Newport')
plt.plot(baltimore_baro, label='Baltimore')
plt.plot(ocean_city_baro, label='Ocean City')
plt.plot(portland_baro, label='Portland')
plt.plot(providence_baro, label='Providence')
plt.axhline(cumm['baro'].mean(), color='r')
plt.xlabel('March 2011 - December 2015', fontsize=30)
plt.ylabel('Barometric Pressure in Millibars (MB)', fontsize=30)
plt.xlim(pd.Timestamp('2011-03-31'), pd.Timestamp('2015-12-31'), )
plt.legend(fontsize ='xx-large')
plt.title(label='Global Air Pressure Average = 1013 MB', 
          fontsize=40);
```
